# Linear Regression

This notebook extends the case seen on binary data for Bernoulli distributions to more complicated data. We will look a **regression task**, a model that relates two continuous variables to eachother. We will simplify things by limiting the hypothesis space and only look at linear relationships. The concepts covered in this notebook are *exaclty* the same as in the binary case. 


We are interested in the task of observing data-set $D=\{\tilde{x}_i, \tilde{y}_i\}^N_{i=1}$ where we assume the following relationship between the variates:

\begin{equation}
y_i = f(\tilde{x}_i)
\end{equation}

Our task is to infer the function $f(\cdot)$ from $D$. To simplify, we will limit the hypothesis space to be only of linear functions. This means we can write the above equation as:

\begin{equation}
y_i = w_1\tilde{x}_i + w_0 = \mathbf w^T\mathbf x_i = \Big[ \Big]
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_line(ax, w):
    # input data
    X = np.zeros((2, 2))
    X[0, 0] = -5
    X[1, 0] = 5
    X[:, 1] = 1
    
    # because of the concatenation we have to flip the transpose
    y = w.dot(X.T)
    ax.plot(X[:, 0], y)

    # create prior distribution
tau = 1.0*np.eye(2)
w_0 = np.zeros((2, 1))

# samples from prior
n_samples = 100

w_samp = np.random.multivariate_normal(w_0.flatten(), tau, size=n_samples)

# creat plot
fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(111)

for i in range(0, w_samp.shape[0]):
    plot_line(ax, w_samp[i, :])
    
plt.show()
```
